<h5>Inicialmente, importamos os dados do link disponilizado na atividade. </h5>

```python
import pandas as pd
url ="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
dados = pd.read_csv(url)
dados.info()
```

Após consultar o dicionário e observar as variáveis e os dados, cheguei a conclusão de que as as variáveis qualitativas são:
<ul>
   <li>Survived</li>
   <li>pclass</li>
   <li>Sex</li>
   <li>embarked</li>
</ul>

Abaixo estão os gráficos de barra e de pizza das respectivas variáveis qualitativas mostradas a cima.
<center><H6>Survived</H6></center>

```python
from matplotlib import pyplot as plt
%matplotlib inline
dados["Survived"].value_counts().plot.bar(title="Relação de Vidas no Titanic", color="#D2B48C")
plt.ylabel("Número de Pessoas")
plt.annotate('Não Sobreviventes', xy=(0.26,340),
             arrowprops=dict(arrowstyle='<-'), xytext=(0.29, 400))
plt.annotate('Sobreviventes', xy=(1,340),
             arrowprops=dict(arrowstyle='<-'), xytext=(1, 390))
```

```python
dados["Survived"].value_counts().plot.pie(title="Relação de Vidas no Titanic")
```

<center><H6>Pclass</H6></center>

```python
dados["Pclass"].value_counts().plot.bar(title="Relação de Classes de Bilhete", color="#FFA07A")
plt.ylabel("Número de Pessoas")
plt.xlabel("Classes")
```

```python
dados["Pclass"].value_counts().plot.pie(title="Relação de Classes de Bilhete")
```

<center><H6>Sex</H6></center>

```python
dados["Sex"].value_counts().plot.bar(title="Relação de homens e mulheres", color="#DEB887")
plt.ylabel("Número de Pessoas")
plt.xlabel("Sexo")
```

```python
dados["Sex"].value_counts().plot.pie()
```

<center><h6>Embarked</h6></center>

```python
dados["Embarked"].value_counts().plot.bar(title="Relação de Locais de Embarcação", color="#A0522D")
plt.ylabel("Número de Pessoas")
plt.xlabel("Local")
```

```python
dados["Embarked"].value_counts().plot.pie()
```
